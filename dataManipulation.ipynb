```python
## filling missing values:

df_filled=df.fillna(0)
print(df_filled)
```

```python
## filling missing values with the mean of column:

df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
df
```

## RENAMING COLUMNS:

```python
df=df.rename(columns={'Date':'Sales date'})
df.head()
```

```python
df=df.rename(columns={'Sale date':'Date'})
df.head()
```

## CHANGE DATATYPE:

```python
df['Value_new']=df['Value'].astype(float)
df.head()
```

```python
df['Value_new']=df['Value'].astype(int)
df.head()
```

```python
df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
## making changes into a particular column

df['Value_new']=df['Value'].apply(lambda x:x*2)    ## value age x chilo ekhn new column value new a x er double thakbe...
df.head()
```

## DATA AGGREGATING AND GROUPING:

```python
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```

```python
grouped_mean=df.groupby(['Product','Region'])['Value'].mean()
print(grouped_mean)
```

```python
## aggregate multiple functions:

grouped_agg=df.groupby('Region')['Value'].agg(['sum','mean','count'])
print(grouped_agg)
```

```python
## merging and joining dataframes:
import pandas as pd

df1=pd.DataFrame({'key':['A','B','C'],'Value':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'Value':[4,5,6]})

```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",how="inner")
```

```python
pd.merge(df1,df2,on="key",how="outer")
```

```python
pd.merge(df1,df2,on="key",how="left")
```

```python
pd.merge(df1,df2,on="key",how="right")
```
