## CREATE A LINE PLOT

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
plt.show()
```

## MULTIPLE PLOTS:

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y2,x,color='yellow')
plt.title("Plot 4")

plt.tight_layout()  ## age title ar x axis er lekha mishe jacchilo otake remove korar jonno
plt.show()


```

## PIE CHARTS:

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']


plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
```

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0) ## move out the 1st slice

plt.pie(sizes,labels=labels,colors=colors,explode=explode,autopct="%1.1f%%")
```

## SALES DATA VISUALIZATION:

```python
import pandas as pd
```

```python
sales_data_df=pd.read_csv('sales_data.csv')
```

```python
sales_data_df.head(5)
```

```python
sales_data_df.info()
```

```python
total_sales_by_products=sales_data_df.groupby('Category')['Sales'].sum()
print(total_sales_by_products)
```

```python
total_sales_by_products.plot(kind='bar',color='teal')
```

```python
## plot sales trend over time

sales_trend=sales_data_df.groupby('Date')['Sales'].sum()
sales_trend
```

```python
sales_trend=sales_data_df.groupby('Date')['Sales'].sum().reset_index()
sales_trend
```

```python
sales_trend=sales_data_df.groupby('Date')['Sales'].sum().reset_index()
sales_trend
plt.plot(sales_trend['Date'],sales_trend['Sales'])
```
