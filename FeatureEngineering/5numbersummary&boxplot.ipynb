```python
import numpy as np
```

```python
lst_marks=[12,45,2,13,90,33,78,6,22,97,54,5,99,17,19,38,43,55,73,84,88,100]
minimum,Q1,median,Q3,maximum=np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum
```

```python
maximum
```

```python
Q1,median,Q3
```

```python
##IQR

IQR=Q3-Q1
print(IQR)
```

```python
## LOWER FENCE & HIGHER FENCE

lower_fence=Q1-1.5*(IQR)
print(lower_fence)

higher_fence=Q3+1.5*(IQR)
print(higher_fence)
```
