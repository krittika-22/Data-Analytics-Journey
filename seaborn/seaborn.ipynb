```python
import seaborn as sns
```

## BASIC PLOTTING WITH SEABORN:

```python
##load data set
tips=sns.load_dataset('tips')
tips
```

```python
## create a scatter plot:

import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("TOTAL BILL VS TIP")
plt.show()
```

```python
## LINE PLOT:

sns.lineplot(x='size',y='total_bill',data=tips)
sns.title("LINE PLOT OF TOTAL BILL VS SIZE")
sns.show()
```

```python
## BAR PLOT:

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of total bill vs day")

plt.show()
```

```python
## BOX PLOT:

sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot of total bill vs day")

plt.show()
 
```

```python
## VIOLIN PLOT:

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin plot of total bill vs day")

plt.show()
```

```python
## HISTOGRAM:

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False)
```

## KDE PLOT:

```python
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## PAIR PLOT:

sns.pairplot(tips)
```

```python
import pandas as pd
sales_data_df=pd.read_csv('sales_data.csv')
sales_data_df.head(5)
```

```python
##plot total sales by product:

plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Sales',data=sales_data_df,estimator=sum)
plt.title("sales by product")
plt.xlabel('product')
plt.ylabel('sales')
plt.show()

```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Sales',data=sales_data_df,estimator=sum)
plt.title("sales by Region")
plt.xlabel('Region')
plt.ylabel('sales')
plt.show()
```
